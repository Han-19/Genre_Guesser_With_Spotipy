```python
def timeit(method):
    def timed(args, **kw):
        ts = time.time()
        result = method(args, **kw)
        te = time.time()
        print ('{}  {:03.2f} ms'.format(method, (te - ts) * 1000))
        return result
    return timed
```

```python
#genre column concatter
def genre_concat(df,genre):
    n = len(df["name"])
    gen=[]
    for i in range(n):
        gen.append(genre)
    return pd.concat([df,pd.DataFrame(gen, columns = ["genre"])], axis=1)
```

```python
for i in X.columns:
    
    ort = np.mean(X[i])
    
    X[i].fillna(ort, inplace=True)
    
    maximum = X[i].max()
    minimum = X[i].min()
    
    
    X[i] = X[i].apply(lambda x: (np.array(x)-minimum)/(maximum-minimum))

X.head()
```

```python
y=df["genre"]
```

```python
#Han
def conf_mat(y_true,y_pred):
    tp = 0
    fp = 0
    tn = 0
    fn = 0
    for i in range(len(y_true)):
        if y_true[i]:
            if(y_pred[i]==y_true[i]):
                tp+=1
            else:
                fp+=1
        else:
            if(y_pred[i]==y_true[i]):
                tn+=1
            else:
                fn+=1
    
    
    cm = [[tp,fn],
          [fp,tn]]
    return cm
############################################
def performance_metrics(cm,yazdır=False):
    tp=cm[0][0]
    tn=cm[1][1]
    fp=cm[1][0]
    fn=cm[0][1]
    
    try:
        accuracy=(tn+tp)/(tp+tn+fn+fp)
    except ZeroDivisionError:
        print("Accuracy hesaplanamadı")
        accuracy=None
    try:
        precision= tp/(tp+fp)
    except ZeroDivisionError:
        print("Precision hesaplanamadı")
        precision=None
    try:
        tpr=tp/(tp+fn)
    except ZeroDivisionError:
        print("Tpr hesaplanamadı")
        tpr=None
    try: 
        fpr=fp/(tn+fp)
    except ZeroDivisionError:
        print("Fpr hesaplanamadı")
        fpr=None
    
    #iş kolaylığı için değiştirdim
    if yazdır:
        print(" accuracy={}\n precision={}\n tpr={}\n fpr={}\n \n".format(accuracy, precision, tpr, fpr))
    return accuracy, precision, tpr, fpr
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
#List for compairing performance records
Perform= []
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
@timeit
def predict_gnb(X_train,X_test,y_train,y_test):
    
    clf_gnb = GaussianNB()
    clf_gnb.fit(X_train,y_train)
    
    y_proba = clf_gnb.predict_proba(X_test)
    y_pred = clf_gnb.predict(X_test)
    y_true = y_test
    
    conf_mat=confusion_matrix(y_test,y_pred)
    
    return (y_true,y_pred,y_proba,conf_mat)
```

```python
y_true,y_pred,y_proba,conf_mat = predict_gnb(X_train, X_test, y_train, y_test)
```

```python
a,b,c,d = performance_metrics(conf_mat,yazdır=True)
Perform.append(["GaussianNB",a,b,c,d])
```

```python
#sklearn.svm.SVC
from sklearn.svm import SVC
def predict_svm(X_train,X_test,y_train,y_test,kernel='linear',prob=True):
    
    clf = SVC(kernel=kernel,probability=prob)
    clf.fit(X_train,y_train)
    
    y_proba =clf.predict_proba(X_test)
    y_pred=clf.predict(X_test)
    
    conf_mat=confusion_matrix(y_test,y_pred)
    y_true = y_test
    
    if prob:
        y_proba = clf.predict_proba(X_test)
        return (y_true,y_pred,y_proba,conf_mat)
    else:
        return (y_true,y_pred,conf_mat)
```

```python
kernel_type = ["poly", "rbf", "sigmoid"]

for i in kernel_type:
    
    y_true,y_pred,y_proba,conf_mat= predict_svm(X_train,X_test,y_train,y_test,kernel=i,prob=True)
    accuracy, precision, tpr, fpr = performance_metrics(conf_mat)
    
    
    print("Kernel type : ",i)
    print("accuracy = ", accuracy, "\n"
     "precision = ", precision, "\n"
     "tpr = ", tpr, "\n"
     "fpr = ", fpr, "\n")
    
    Perform.append(["SVC {} type".format(i),accuracy,precision,tpr,fpr])
```

```python
#sklearn.tree.DecisionTreeClassifier
#from ... import ...
from sklearn.tree import DecisionTreeClassifier

def predict_dt(X_train,X_test,y_train,y_test):
    
    clf_dt = DecisionTreeClassifier()
    clf_dt.fit(X_train,y_train)
    
    y_proba = clf_dt.predict_proba(X_test)
    y_pred = clf_dt.predict(X_test)
    
    conf_mat=confusion_matrix(y_test,y_pred)
    y_true = y_test
    return (y_true,y_pred,y_proba,conf_mat)
```

```python
y_true,y_pred,y_proba,conf_mat = predict_dt(X_train,X_test,y_train,y_test)
accuracy, precision, tpr, fpr = performance_metrics(conf_mat)
print("accuracy = ", accuracy, "\n"
     "precision = ", precision, "\n"
     "tpr = ", tpr, "\n"
     "fpr = ", fpr, "\n")
Perform.append(["DecisionTree",accuracy,precision,tpr,fpr])
```

```python
#sklearn.neighbors.KNeighborsClassifier
#from ... import ...
from sklearn.neighbors import KNeighborsClassifier

def predict_knn(X_train,X_test,y_train,y_test,k):

    clf_knn = KNeighborsClassifier(n_neighbors=k)
    clf_knn.fit(X_train,y_train)
    
    y_proba =clf_knn.predict_proba(X_test)
    y_pred=clf_knn.predict(X_test)
    
    conf_mat=confusion_matrix(y_test,y_pred)
    y_true = y_test
    
    return (y_true,y_pred,y_proba,conf_mat)
```

```python
k = range(1,20,2)
for i in k:
    y_true,y_pred,y_proba,conf_mat= predict_knn(X_train,X_test,y_train,y_test,i)
    
    accuracy, precision, tpr, fpr = performance_metrics(conf_mat)
   
    print("k = ",i)
    print("accuracy = ", accuracy, "\n"
     "precision = ", precision, "\n"
     "tpr = ", tpr, "\n"
     "fpr = ", fpr, "\n")
    Perform.append(["Kneighbours k= {} ".format(i),accuracy,precision,tpr,fpr])
```

```python
error = []
# Calculating error for K values between 1 and 7
for i in range(1, 40):
    y_true,y_pred,y_proba,conf_mat = predict_knn(X_train,X_test,y_train,y_test,i)

    error.append(np.mean(y_pred != y_test))

plt.figure(figsize=(12, 6))
plt.plot(range(1, 40), error, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
```
